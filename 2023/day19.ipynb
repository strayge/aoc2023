```python
rules = {}
workflows = []

wf_started = False
for line in lines:
    if not line:
        wf_started = True
        continue
    if wf_started:
        wf = {}
        parts = line[1:-1].split(',')
        for part in parts:
            k, v = part.split('=', 1)
            wf[k] = int(v)
        workflows.append(wf)
    else:
        name, rule_str = line[:-1].split('{', 1)
        rule = []
        rule_parts = rule_str.split(',')
        for part in rule_parts:
            if ':' not in part:
                rule.append(('', '', 0, part))
                continue
            part1, target = part.split(':', 1)
            rule.append((part1[0], part1[1], int(part1[2:]), target))
        rules[name] = rule

# print(rules)
# print(workflows)
```

```python
result = 0
for wf in workflows:
    target = 'in'
    while True:
        rule = rules[target]
        for rule_var, rule_sign, rule_value, rule_target in rule:
            if not rule_var:
                target = rule_target
                break
            if rule_sign == '<' and wf[rule_var] < rule_value:
                target = rule_target
                break
            if rule_sign == '>' and wf[rule_var] > rule_value:
                target = rule_target
                break
        if target == 'R':
            break
        if target == 'A':
            result += sum(wf.values())
            break

print(result)
```

```python
send_result(DAY, 1, result)
```

```python
from copy import deepcopy


def vars_to_str(vars: dict[str, list[int]]) -> str:
    parts = []
    for var, nums in vars.items():
        min_var, max_var = nums
        parts.append(f'{var}:{min_var}-{max_var}')
    return '[' + ', '.join(parts) + ']'

def rules_to_str(rules_list: list[tuple[str, str, int, str]]) -> str:
    parts = []
    for rule_var, rule_sign, rule_value, rule_target in rules_list:
        parts.append(f'{rule_var}{rule_sign}{rule_value}:{rule_target}')
    return '{' +  ', '.join(parts) + '}'

def calc_perms(vars: dict[str, list[int]]) -> int:
    for var, nums in vars.items():
        min_var, max_var = nums
        if min_var >= max_var:
            return 0
    result = (vars['x'][1] - vars['x'][0]) * (vars['m'][1] - vars['m'][0]) * (vars['a'][1] - vars['a'][0]) * (vars['s'][1] - vars['s'][0])
    # print(f'calc A => {result}')
    return result


def calculate_combinations(rules_list: list[tuple[str, str, int, str]], vars: dict[str, list[int]], deep: int = 0) -> int:
    deep_str = '    ' * deep
    print(f'{deep_str}call {rules_to_str(rules_list)}  {vars_to_str(vars)}')
    for var, nums in vars.items():
        min_var, max_var = nums
        if min_var >= max_var:
            print(f'{deep_str}stop {var} {min_var} >= {max_var} => 0')
            return 0

    result = 0
    for rule_var, rule_sign, rule_value, rule_target in rules_list:
        print(f'{deep_str}step {{{rule_var}{rule_sign}{rule_value}:{rule_target}}}  {vars_to_str(vars)}')
        if not rule_var:
            if rule_target == 'R':
                print(f'{deep_str}stop R => {result}')
                return result
            if rule_target == 'A':
                sub_result = calc_perms(vars)
                print(f'{deep_str}stop A {vars_to_str(vars)} => + {sub_result} = {result + sub_result}')
                result += sub_result
                return result
            print(f'{deep_str}new call for "" {rule_target}  {vars_to_str(vars)}')
            sub_result = calculate_combinations(rules[rule_target], vars, deep=deep + 1)
            print(f'{deep_str}return from "" {rule_target}  {vars_to_str(vars)} => + {sub_result} = {result + sub_result}')
            result += sub_result
            return result

        if rule_sign == '<':
            new_vars = deepcopy(vars)
            new_vars[rule_var][1] = min(new_vars[rule_var][1], rule_value)
            if rule_target == 'R':
                print(f'{deep_str}pass R')
                pass
            elif rule_target == 'A':
                sub_result = calc_perms(new_vars)
                print(f'{deep_str}add A {vars_to_str(vars)} => + {sub_result} = {result + sub_result}')
                result += sub_result
            else:
                print(f'{deep_str}new call for "<" {rule_target}  {vars_to_str(new_vars)}')
                sub_result = calculate_combinations(rules[rule_target], new_vars, deep=deep + 1)
                print(f'{deep_str}return from "<" {rule_target}  {vars_to_str(new_vars)} => + {sub_result} = {result + sub_result}')
                result += sub_result
            vars[rule_var][0] = max(vars[rule_var][0], rule_value)
            print(f'{deep_str}vars after "<" {vars_to_str(vars)}')

        if rule_sign == '>':
            new_vars = deepcopy(vars)
            new_vars[rule_var][0] = max(new_vars[rule_var][0], rule_value + 1)

            if rule_target == 'R':
                print(f'{deep_str}pass R')
                pass
            elif rule_target == 'A':
                sub_result = calc_perms(new_vars)
                print(f'{deep_str}add A {vars_to_str(vars)} => + {sub_result} = {result + sub_result}')
                result += sub_result
            else:
                print(f'{deep_str}new call for ">" {rule_target}  {vars_to_str(new_vars)}')
                sub_result = calculate_combinations(rules[rule_target], new_vars, deep=deep + 1)
                print(f'{deep_str}return from ">" {rule_target}  {vars_to_str(new_vars)} => + {sub_result} = {result + sub_result}')
                result += sub_result

            vars[rule_var][1] = min(vars[rule_var][1], rule_value + 1)
            print(f'{deep_str}vars after ">" {vars_to_str(vars)}')

    print(f'{deep_str}call end   {rules_to_str(rules_list)}   {vars_to_str(vars)}  =>  {result}')
    return result


result = calculate_combinations(rules['in'], {'x': [1, 4001], 'm': [1, 4001], 'a': [1, 4001], 's': [1, 4001]})
print(result)
```

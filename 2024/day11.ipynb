```python
def num_change(num: int) -> list[int]:
    if num == 0:
        return [1]
    num_str = str(num)
    if len(num_str) % 2 == 0:
        num1 = int(num_str[:len(num_str)//2])
        num2 = int(num_str[len(num_str)//2:])
        return [num1, num2]
    return [num * 2024]
```

```python
next_numbers = numbers
for i in range(25):
    cur_numbers = next_numbers[:]
    next_numbers = []
    for num in cur_numbers:
        next_numbers.extend(num_change(num))
    # print(next_numbers)
print(len(next_numbers))
```

```python
numbers1 = []
for x in numbers:
    numbers1.extend(num_change(x))

cached37nums: dict[int, list[int]] = {}
cached37counts: dict[int, int] = {}

def cache_num(num: int) -> None:
    if num in cached37nums:
        return
    prev_nums = [num]
    for i in range(37):
        new_nums = []
        for prev_num in prev_nums:
            new_nums.extend(num_change(prev_num))
        prev_nums = new_nums
    cached37nums[num] = prev_nums

def cache_count(num: int) -> None:
    if num in cached37counts:
        return
    if num in cached37nums:
        cached37counts[num] = len(cached37nums[num])
        return
    prev_nums = [num]
    for i in range(37):
        new_nums = []
        for prev_num in prev_nums:
            new_nums.extend(num_change(prev_num))
        prev_nums = new_nums
    cached37counts[num] = len(prev_nums)


# precalc: dict[int, list[list[int]]] = {}
# def add_precalc(num: int, iter: int) -> None:
#     if num in precalc and len(precalc[num]) > iter:
#         return
#     print('calc:', num)
#     precalc[num] = [[num]]
#     for i in range(1, iter + 1):
#         prev_nums = precalc[num][i-1]
#         new_nums = []
#         for prev_num in prev_nums:
#             new_nums.extend(num_change(prev_num))
#         precalc[num].append(new_nums)

for num in numbers1:
    cache_num(num)
```

```python
half_nums = []
for num in numbers1:
    half_nums.extend(cached37nums[num])
print(len(set(half_nums)))
```

```python
m = Map.from_object(lines)
m.show()
```

```python
m_processed = Map(m.h, m.w)
groups = {}
i = 0
while True:
    i += 1
    points = []
    try:
        y, x = m_processed.find_value(-1)
        value = m[y][x]
    except ValueError:
        break

    def target_filter(y1, x1, value1, y2, x2, value2):
        return value1 == value and value2 == value and m_processed[y2][x2] == -1

    def action(y1, x1, value1, y2, x2, value2):
        m_processed[y2][x2] = i
        points.append((y2, x2))
        print()
        m_processed.show()

    m.calc_step(target_filter=target_filter, target_action=action)
    groups[i] = points


m_processed.show()
```

```python
result = 0
for group_num, points in groups.items():
    square = len(points)
    perimeter = 0
    for y, x in points:
        nearest = list(m_processed.get_nearest_points_by_value(y, x, group_num))
        print(f'{group_num}. {y, x} {len(nearest)} => {4 - len(nearest)}')
        perimeter += 4 - len(nearest)
    print(f'group {group_num}: square={square}, perimeter={perimeter}')
    result += square * perimeter
    # print(group_num, len(points))
print(result)
```

```python
send_result(DAY, 1, result)
```

```python
def combo(n):
    global A, B, C
    if n <= 3:
        return n
    if n == 4:
        return A
    if n == 5:
        return B
    if n == 6:
        return C

def adv(n):
    global A
    A = A // (2 ** combo(n))

def bxl(n):
    global B
    B = B ^ n

def bst(n):
    global B
    B = combo(n) % 8

def bxc(n):
    global B
    B = B ^ C

def out(n):
    print(f'{combo(n) % 8},', end='')

def bdv(n):
    global B
    B = A // (2 ** combo(n))

def cdv(n):
    global C
    C = A // (2 ** combo(n))

def jnz(n):
    global A
    return A == 0


A = 50230824
B = 0
C = 0
while True:
    bst(4)
    bxl(3)
    cdv(5)
    adv(3)
    bxl(4)
    bxc(7)
    out(5)
    if jnz(0):
        break
```

```python
def one_iteration(A_high_raw):
    A_low = A_high_raw % 8 ^ 3
    A_high = A_high_raw >> A_low
    # print(A_low, A_high)
    return (A_low ^ 4 ^ A_high) % 8

A = 50230824
while A > 0:
    print(one_iteration(A % 512), end=',')
    A = A // 8
```

```python
def step(A, program):
    program_nums = [int(x) for x in program]
    B = 0
    C = 0
    output = ''

    def combo(n) -> int:
        if n <= 3:
            return n
        if n == 4:
            return A
        if n == 5:
            return B
        if n == 6:
            return C

    for i in range(0, len(program_nums), 2):
        instr = program_nums[i]
        n = program_nums[i + 1]
        match instr:
            case 0:  # adv
                A = A // (2 ** combo(n))
            case 1:  # bxl
                B = B ^ n
            case 2:  # bst
                B = combo(n) % 8
            case 3:  # jnz
                return A, output
            case 4:  # bxc
                B = B ^ C
            case 5:  # out
                output += str(combo(n) % 8)
            case 6:  # bdv
                B = A // (2 ** combo(n))
            case 7:  # cdv
                C = A // (2 ** combo(n))

results = []
def step2(pos, full_val, program):
    line = program[::-1]
    if pos >= len(line):
        print(full_val)
        results.append(full_val)
        return
    expected = line[pos]
    for x in range(8):
        new_val = (full_val << 3) + x
        _, out = step(new_val, program)
        if out == expected:
            step2(pos + 1, new_val, program)


# step2(0, 0, '035430')
step2(0, 0, '2413750314475530')

print('min', min(results))
```

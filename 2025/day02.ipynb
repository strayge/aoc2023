```python
lines = get_input_lines(DAY)
```

```python
ranges = []
for line in lines[0].split(','):
    num1, num2 = line.split('-')
    ranges.append((int(num1), int(num2)))

count = 0
result = 0
for start, end in ranges:
    for i in range(start, end + 1):
        i_str = str(i)
        length = len(i_str)
        for split_by in range(1, length // 2 + 1):
            if length % split_by != 0:
                continue
            parts_count = length // split_by
            parts = [i_str[i * split_by:(i + 1) * split_by] for i in range(parts_count)]
            if len(set(parts)) == 1:
                # print(f'{start}-{end}: {i} -> {parts}')
                count += 1
                result += i
                break

count, result
```

```python
send_result(DAY, 2, result)
```

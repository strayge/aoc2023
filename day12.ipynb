```python
from itertools import product

def count_line(line: str) -> list[int]:
    nums = []
    length = 0
    for char in line:
        if char == '#':
            length += 1
            continue
        if length > 0:
            nums.append(length)
        length = 0
    if length > 0:
        nums.append(length)
    return nums

def get_line_variants(line: str) -> list[list[str]]:
    variants = []
    for char in line:
        if char == '?':
            variants.append(['.','#'])
        else:
            variants.append([char])
    return variants

total_counter = 0
for line, correct_nums in lines2:
    variants = get_line_variants(line)
    counter = 0
    for possible_line in product(*variants):
        possible_nums = count_line(possible_line)
        if possible_nums == correct_nums:
            counter += 1
    total_counter += counter
    # print(line, counter)
print(total_counter)
```

```python
def unfold_line(line: str, nums: list[int], repeat: int = 5) -> tuple[str, list[int]]:
    new_line = '?'.join([line] * repeat)
    new_nums = nums * repeat
    return new_line, new_nums
```

```python
def count_correct_variants(line: str, correct_nums: list[int], timeout: float = 0) -> int | None:
    start_time = time.monotonic()
    variants = get_line_variants(line)
    counter = 0
    for possible_line in product(*variants):
        if timeout and time.monotonic() - start_time > timeout:
            return None
        possible_nums = count_line(possible_line)
        if possible_nums == correct_nums:
            counter += 1
    return counter

for line, nums in lines2:
    print(f'line: {line:20} nums: {nums!s:20}')
    for i in range(1, 5):
        new_line, new_nums = unfold_line(line, nums, i)
        count = count_correct_variants(new_line, new_nums, timeout=20)
        if count is None:
            break
        print(f'{i:3} {count:5}')
```

```python
def make_partial_sharps_up_to_min_length(line: str, nums: list[int]) -> str:
    min_num = min(nums)
    parts = line.split('.')
    for i, part in enumerate(parts):
        if '#' not in part:
            continue
        if part.startswith('#'):
            parts[i] = '#' * min_num + part[min_num:]
        if part.endswith('#'):
            parts[i] = part[:-min_num] + '#' * min_num
    line = '.'.join(parts)
    return line

def filter_by_max_length(lines: list[str], nums: list[int]) -> list[str]:
    result = []
    for line in lines:
        max_allowed_part = len(max(line.split('.'), key=len))
        min_allowed_part = len(max(line.replace('?', '.').split('.'), key=len))
        if max_allowed_part < max(nums):
            continue
        if min_allowed_part > min(nums):
            continue
        result.append(line)
    return result

def get_max_sharp_variants(line: str, nums: list[int], deep=0) -> set[str]:
    # print(f'get_max_sharp_variants {deep=} start {line=} {nums=}')
    if not nums:
        if '#' in line:
            # print(f'get_max_sharp_variants {deep=} end1 {line=} {nums=} []')
            return set()
        # print(f'get_max_sharp_variants {deep=} end2 {line=} {nums=} ["{line.replace("?", ".")}"]')
        return {line.replace('?', '.')}

    max_number = max(nums)
    max_number_pos = nums.index(max_number)  # get first one

    nums_before = nums[:max_number_pos]
    nums_after = nums[max_number_pos+1:]
    len_before = sum(nums_before) + len(nums_before)
    len_after = sum(nums_after) + len(nums_after)

    variants = set()
    for i in range(len_before, len(line) - len_after):
        substring = line[i:i+max_number]
        string_before = line[:i]
        string_after = line[i+max_number:]
        char_before = string_before[-1] if string_before else None
        char_after = string_after[0] if string_after else None
        if '.' in substring:
            # print(f'get_max_sharp_variants {deep=} skip1 {line=} {nums=} {i=} "{string_before}  {substring}  {string_after}"')
            continue
        if char_before and char_before == '#':
            # print(f'get_max_sharp_variants {deep=} skip2 {line=} {nums=} {i=} "{string_before}  {substring}  {string_after}"')
            continue
        if char_after and char_after == '#':
            # print(f'get_max_sharp_variants {deep=} skip3 {line=} {nums=} {i=} "{string_before}  {substring}  {string_after}"')
            continue

        if char_before:
            string_before = string_before[:-1] + '.'
        if char_after:
            string_after = '.' + string_after[1:]
        substring = '#' * max_number

        variants_before = get_max_sharp_variants(string_before, nums_before, deep+1) if string_before else [string_before]
        variants_after = get_max_sharp_variants(string_after, nums_after, deep+1) if string_after else [string_after]

        # print(f'get_max_sharp_variants {deep=} product {line=} {nums=} {i=} {substring=} {variants_before=} {variants_after=}')
        if not variants_before or not variants_after:
            continue
        for new_variant in product(variants_before, [substring], variants_after):
            new_line = ''.join(new_variant)
            # print(f'get_max_sharp_variants {deep=} add {line=} {nums=} => {new_line}')
            variants.add(new_line)

    return variants

def split_lines_to_variants(line: str, nums: list[int]) -> list[str]:
    # line = make_partial_sharps_up_to_min_length(line, nums)
    variants = get_max_sharp_variants(line, nums)
    return variants

    new_variants = set()
    for variant in variants:
        # parts = list(filter(None, variant.split('.')))
        # # print(variant, parts)
        # part_i = -1
        # part_started = False
        # part_start = 0
        # part_end = 0
        # invalid_variant = False
        # for i, c in enumerate(variant + '.'):
        #     if c == '.' and part_started:
        #         part_started = False
        #         part_end = i
        #         part = variant[part_start:part_end]
        #         correct_len = nums[part_i] if part_i < len(nums) else 0
        #         # correct_part = parts[part_i]
        #         print('ended part', part, 'correct', correct_len)
        #         if len(part) < correct_len:
        #             if '#' not in part:
        #                 # replace part with .
        #                 print(1, variant)
        #                 variant = variant[:part_start] + '.' * (part_end-part_start) + variant[part_end:]
        #                 print(2, variant)
        #             else:
        #                 invalid_variant = True
        #                 break

        #     if c != '.' and not part_started:
        #         part_started = True
        #         part_start = i
        #         part_i += 1
        # if invalid_variant:
        #     continue
        new_variants.add(variant)

    return list(new_variants)

# lines2[0], split_lines_to_variants(*lines2[0])
```

```python
def count_correct_variants_v2(line: str, correct_nums: list[int], timeout: float = 0) -> int | None:
    start_time = time.monotonic()
    max_counter = 0
    for line_variant in split_lines_to_variants(line, correct_nums):
        counter = 0
        variants = get_line_variants(line_variant)
        for possible_line in product(*variants):
            if timeout and time.monotonic() - start_time > timeout:
                return None
            possible_nums = count_line(possible_line)
            if possible_nums == correct_nums:
                counter += 1
        max_counter = max(max_counter, counter)
    # print(line, max_counter)
    return max_counter
```
